```python
import util
import pandas as pd
from collections import namedtuple
```

```python
def get_params():
    hps_dict = {
        'train_file' : '/home/ubuntu/data/summary/train/*.csv',
        'validate_file' : '/home/ubuntu/data/summary/validate/*.csv',
        'test_file' : '/home/ubuntu/data/summary/test/*.csv',
        'loss_func':'mae',
        'optimizer':'adam',
        'hidden_dim' : 250,
        'batch_size' : 400,
        'epochs' : 120,
        'model_lstm':True,
        'filter_outliers':True,
        'upper_limit':99.5,
        'baseline_mae':4520.485,
        'dropout':[0.0,0.1,0.2,0.4,0.6]
    }
    
    hps = namedtuple("HParams", hps_dict.keys())(**hps_dict)
    return hps
```

```python
def run():
    hps = get_params()
    rmse_l = list()
    mae_l = list()
    norm_l = list()
    for dropout in hps.dropout:
        rmse,mae,norm = util.experiment(hps,dropout)
        rmse_l.append(rmse)
        mae_l.append(mae)
        norm_l.append(norm)
    
    print('RESULTS:')
    print('=========')
    i=0
    for dropout in hps.dropout:
        print('Droput: {0} ===> RMSE: {1} ; MAE: {2} ; Normalized MAE: {3}'.format(dropout,
                                                                                   rmse_l[i],
                                                                                   mae_l[i],
                                                                                   norm_l[i]))
        i += 1
    
```

```python
run()
```
